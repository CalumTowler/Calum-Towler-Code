```python
# Packages
import lightkurve as lk
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 1: Load the Kepler-10 light curve
# The dataset will be automatically downloaded if not available locally
lc = lk.search_lightcurve('Kepler-10').download()
print('Step 1 Complete')

# Step 2: Plot the raw light curve
plt.figure(figsize=(10, 5))
lc.plot()
plt.title('Raw Light Curve')
plt.xlabel('Time [BJD]')
plt.ylabel('Flux')
plt.show()
print('Step 2 Complete')

# Step 3: Transit Search
# Adjust the period range for the periodogram
min_period = 1.0  # Minimum period in days
max_period = 10.0  # Maximum period in days
periodogram = lc.to_periodogram(method='bls', period=np.arange(min_period, max_period, 0.001))
transit_period = periodogram.period_at_max_power
print('Step 3 Complete')

# Step 4: Plot the periodogram
plt.figure(figsize=(10, 5))
periodogram.plot()
plt.title('Periodogram')
plt.xlabel('Period [days]')
plt.ylabel('Power')
plt.show()
print('Step 4 Complete')

# Step 5: Plot the folded light curve with detected transit events
plt.figure(figsize=(10, 5))
folded_lc = lc.fold(period=transit_period)
folded_lc.scatter()
plt.title('Folded Light Curve')
plt.xlabel('Phase')
plt.ylabel('Flux')
plt.show()
print('Step 5 Complete')
```
